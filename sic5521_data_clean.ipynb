###Correct misspellings

```python
def clean_columns(df, column, wrong_spell, corrected_spell):

    df[column] = df[column].str.replace(wrong_spell, corrected_spell, regex=False)
    return df
```

```python
#misspelling from "Annual Sales" column:
#$500,000 To 1 Millio
#$500,000 To $1 Million

sic5521_corrected_annual_sales = clean_columns(sic5521_cleaned, "Annual Sales", "$500,000 To 1 Millio", "$500,000 To $1 Million")
sic5521_corrected_annual_sales
```

```python
#misspelling from "Industry" column:
#Used Car Dealers; Nec
#Used Car Dealers, Nec

#Motor Vehicle Dealers (used Only)
#Motor Vehicle Dealers-used Only

sic5521_corrected_industry = clean_columns(sic5521_corrected_annual_sales, "Industry", "Used Car Dealers; Nec", "Used Car Dealers, Nec")
sic5521_corrected_industry = clean_columns(sic5521_corrected_industry, "Industry", "Motor Vehicle Dealers (used Only)", "Motor Vehicle Dealers-used Only")

sic5521_corrected_industry
```

####Phone

```python
def format_phone(phone):
    if (phone) == 'Missing Phone #':
        return 'Missing Phone #'
    phone_str = str(int(phone))  # Convert to integer and then to string to remove scientific notation
    if len(phone_str) == 10:  # Standard 10-digit phone number
        return f"({phone_str[:3]}) {phone_str[3:6]}-{phone_str[6:]}"
    else:
        return 'Missing Phone #'  # Replace non-10-digit numbers with '*'


# Apply the function to the 'Phone' column
sic5521_corrected_sic['Phone'] = sic5521_corrected_sic['Phone'].apply(format_phone)

# Display the result
sic5521_foramted_phone = sic5521_corrected_sic
sic5521_foramted_phone
```

####Zip

```python
def format_zip(zip_code):
    if (zip_code) == 'Missing Zip Code':
        return 'Missing Zip Code'
    elif '-' in zip_code:
        return zip_code.split('-')[0]  # Take the first 5 digits
    elif len(zip_code) == 4:
        return 'Only four digits #:' + zip_code  # Add leading '#' to make it 5 characters
    elif len(zip_code) == 5 and zip_code.isdigit():
        return zip_code  # Valid 5-digit zip code
    else:
        return 'Missing Zip Code'

# Apply the function directly to the 'Zip' column
sic5521_foramted_phone['Zip'] = sic5521_foramted_phone['Zip'].apply(format_zip)

#Ensure the column is treated as text by converting all entries to strings, it keeps the leading 0s when view in Excel
sic5521_foramted_phone['Zip'] = sic5521_foramted_phone['Zip'].astype(str)

# Display the result
sic5521_foramted_zip = sic5521_foramted_phone
sic5521_foramted_zip
```
