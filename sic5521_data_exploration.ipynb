##Data Exploration

```python
#Check data type and null values of each column
def column_describe(df):
    data = []
    for column in df.columns:
      data.append([
                  column, \
                  df[column].dtype, \
                  df[column].notna().sum(), \
                  df[column].isna().sum(), \
                  f"{round(100*(df[column].isna().sum()/len(df)),2)}%", \
                  df[column].nunique()
                  ])
    return pd.DataFrame(columns=['Column Names','Data Type','# of Not Null Value','# of Null Value','Null Value Percentage','# of Unique Value'], data=data)

column_describe(sic5521)
```

##Find out SIC Codes and corresponds Industries

```python
#Count the number of unique SIC Codes and Industry names
num_sic_code = sic5521['SIC Code'].nunique()
num_industry = sic5521['Industry'].nunique()

#List the unique SIC Codes and Industry names
unique_sic_code = sic5521['SIC Code'].unique()
unique_industry = sic5521['Industry'].unique()

print(f"There are {num_sic_code} different SIC Codes")
print(f"There are {num_industry} different Industry names")
print(f"They are: {unique_sic_code}")
print(f"They are: {unique_industry}")

#Showing the distribution of values in the "SIC Code" column
def col_dist(df):
  return df['SIC Code'].value_counts()
col_dist(sic5521)
```

##Find out misspellings

```python
#the function to show the distribution of a certain column
def col_dist(column):
  return sic5521[column].value_counts()

#the function to count how many unique values in a certain column
def uniq_value(column):
  uniq = sic5521[column].nunique()
  return print(f"There are {uniq} different values in the {column} column")

#the function to list out unique values in a certain column(not include null values)
def sorted_unique(column):
  unique = sorted(sic5521[column].dropna().unique())
  return unique
```
